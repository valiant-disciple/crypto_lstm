```python
import os

import glob

# Define the directory where the CSV files were extracted
csv_directory = "/content/drive/MyDrive/updated/public_dataset/train/tweets/AGI"

# Use glob to find all CSV files in the directory
csv_files = glob.glob(os.path.join(csv_directory, "*.csv"))

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each CSV file and load it into a DataFrame
for file in csv_files:
    try:
        # Read the CSV file
        df = pd.read_csv(file)
        dataframes.append(df)
        print(f"Loaded {file}")
    except Exception as e:
        print(f"Failed to load {file}: {e}")

# Concatenate all DataFrames into one if needed
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the first few rows of the combined DataFrame
print(combined_df.head())
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords if needed
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_tweet(tweet):
    # Remove URLs, mentions, hashtags
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    tweet = re.sub(r'\@\w+|\#','', tweet)

    # Remove special characters and numbers
    tweet = re.sub(r'[^A-Za-z\s]', '', tweet)

    # Convert to lowercase and remove stopwords
    tweet = tweet.lower()
    tweet_tokens = tweet.split()
    filtered_words = [word for word in tweet_tokens if word not in stop_words]

    return ' '.join(filtered_words)
```

```python
for idx, row in combined_df.iterrows():
  text=str(row['tweet'])
  cleaned_text = clean_tweet(text)
  combined_df.at[idx, 'tweet'] = cleaned_text
```

```python
import os
coin_data_train = {}
coins = []
folder_path = '/content/drive/MyDrive/updated/public_dataset/train/price_data'
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        coin_name = file_name.split('.')[0]

        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        # df.fillna(method='ffill', inplace=True)
        df.set_index('datetime', inplace=True)
        df.sort_values('datetime', inplace=True)
        coin_data_train[coin_name] = df

coin_data_test = {}
folder_path = '/content/drive/MyDrive/updated/public_dataset/test/price_data'
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        coin_name = file_name.split('.')[0]
        coins.append(coin_name)
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        print(df.head())
        # df.set_index('datetime', inplace=True)
        df.sort_values('datetime', inplace=True)
        # df.fillna(method='ffill', inplace=True)
        df.sort_values('datetime', inplace=True)
        coin_data_test[coin_name] = df

print(coins)
```

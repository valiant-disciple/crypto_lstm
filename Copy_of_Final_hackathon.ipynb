```python
import os

import glob

# Define the directory where the CSV files were extracted
csv_directory = "/content/drive/MyDrive/updated/public_dataset/train/tweets/AGI"

# Use glob to find all CSV files in the directory
csv_files = glob.glob(os.path.join(csv_directory, "*.csv"))

# Initialize an empty list to store DataFrames
dataframes = []

# Loop through each CSV file and load it into a DataFrame
for file in csv_files:
    try:
        # Read the CSV file
        df = pd.read_csv(file)
        dataframes.append(df)
        print(f"Loaded {file}")
    except Exception as e:
        print(f"Failed to load {file}: {e}")

# Concatenate all DataFrames into one if needed
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the first few rows of the combined DataFrame
print(combined_df.head())
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords if needed
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_tweet(tweet):
    # Remove URLs, mentions, hashtags
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    tweet = re.sub(r'\@\w+|\#','', tweet)

    # Remove special characters and numbers
    tweet = re.sub(r'[^A-Za-z\s]', '', tweet)

    # Convert to lowercase and remove stopwords
    tweet = tweet.lower()
    tweet_tokens = tweet.split()
    filtered_words = [word for word in tweet_tokens if word not in stop_words]

    return ' '.join(filtered_words)
```

```python
import os
import glob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_scores(sentence):
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)
    return sentiment_dict['compound']

def per_coin_sentiment(coin):
  # Define the directory where the CSV files were extracted
  tweet_folder = "/content/drive/MyDrive/updated/public_dataset/train/tweets/"
  csv_directory = os.path.join(tweet_folder, coin)

  # Use glob to find all CSV files in the directory
  csv_files = glob.glob(os.path.join(csv_directory, "*.csv"))

  # Initialize an empty list to store DataFrames
  dataframes = []

  # Loop through each CSV file and load it into a DataFrame
  for file in csv_files:
      try:
          # Read the CSV file
          df = pd.read_csv(file)
          dataframes.append(df)
      except Exception as e:
          print(f"Failed to load {file}: {e}")

  # Concatenate all DataFrames into one if needed
  combined_df = pd.concat(dataframes, ignore_index=True)
  combined_df = combined_df[['date', 'tweet']]

  sentiments = []
  tweets = combined_df['tweet']
  for tweet in tweets:
      sentiment_score = sentiment_scores(str(tweet))
      sentiments.append(sentiment_score)

  return sentiments
```
